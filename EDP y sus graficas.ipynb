```
import numpy as np
import IPython.display as display
from matplotlib import pyplot as plt
import io
import base64

ys = 200 + np.random.randn(100)
x = [x for x in range(len(ys))]

fig = plt.figure(figsize=(4, 3), facecolor='w')
plt.plot(x, ys, '-')
plt.fill_between(x, ys, 195, where=(ys > 195), facecolor='g', alpha=0.6)
plt.title("Sample Visualization", fontsize=10)

data = io.BytesIO()
plt.savefig(data)
image = F"data:image/png;base64,{base64.b64encode(data.getvalue()).decode()}"
alt = "Sample Visualization"
display.display(display.Markdown(F"""![{alt}]({image})"""))
plt.close(fig)
```

```
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

# Variables simbólicas
x, lam = sp.symbols('x lambda', real=True)
X = sp.Function('X')

# Soluciones simbólicas
n = sp.symbols('n', integer=True, positive=True)
lambda_n = (n * sp.pi)**2

# Funciones propias
Xn_dirichlet = sp.sin(n * sp.pi * x)
Xn_neumann = sp.cos(n * sp.pi * x)

# Lambdify para graficar (crear funciones evaluables numéricamente)
Xn_dirichlet_fns = [sp.lambdify(x, Xn_dirichlet.subs(n, i), 'numpy') for i in range(1, 4)]
Xn_neumann_fns  = [sp.lambdify(x, Xn_neumann.subs(n, i), 'numpy') for i in range(1, 4)]

# Valores de x para graficar
x_vals = np.linspace(0, 1, 200)

# Graficar funciones propias (Dirichlet)
plt.figure(figsize=(10, 4))
for i, f in enumerate(Xn_dirichlet_fns, 1):
    plt.plot(x_vals, f(x_vals), label=f"n={i}, sin({i}πx)")
plt.title("Funciones propias - Condiciones Dirichlet")
plt.xlabel("x")
plt.ylabel("X_n(x)")
plt.legend()
plt.grid(True)
plt.show()

# Graficar funciones propias (Neumann)
plt.figure(figsize=(10, 4))
for i, f in enumerate(Xn_neumann_fns, 1):
    plt.plot(x_vals, f(x_vals), label=f"n={i}, cos({i}πx)")
plt.title("Funciones propias - Condiciones Neumann")
plt.xlabel("x")
plt.ylabel("X_n(x)")
plt.legend()
plt.grid(True)
plt.show()
```

```
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_bvp

# Definir el sistema de ecuaciones diferenciales
# y[0] = X, y[1] = X'
def odes(x, y, lambda_n):
    return np.vstack([y[1], -lambda_n * y[0]])

# Condiciones de frontera para Dirichlet
def bc_dirichlet(ya, yb):
    return np.array([ya[0], yb[0]])

# Condiciones de frontera para Neumann
def bc_neumann(ya, yb):
    return np.array([ya[1], yb[1]])

# Valores de x para la solución numérica
x_vals = np.linspace(0, 1, 100)

# Solución numérica para Dirichlet
n_values = [1, 2, 3]
dirichlet_solutions = []
for n in n_values:
    lambda_n = (n * np.pi) ** 2
    y_init = np.zeros((2, x_vals.size))  # Condición inicial aproximada
    sol = solve_bvp(lambda x, y: odes(x, y, lambda_n), bc_dirichlet, x_vals, y_init)
    dirichlet_solutions.append(sol.sol(x_vals)[0])

# Solución numérica para Neumann
neumann_solutions = []
for n in n_values:
    lambda_n = (n * np.pi) ** 2
    y_init = np.zeros((2, x_vals.size))  # Condición inicial aproximada
    sol = solve_bvp(lambda x, y: odes(x, y, lambda_n), bc_neumann, x_vals, y_init)
    neumann_solutions.append(sol.sol(x_vals)[0])

# Graficar soluciones numéricas para Dirichlet
plt.figure(figsize=(12, 6))
for i, sol in enumerate(dirichlet_solutions):
    plt.plot(x_vals, sol, label=f"n={n_values[i]}, Numérico")
    plt.plot(x_vals, np.sin(n_values[i] * np.pi * x_vals), '--', label=f"n={n_values[i]}, Analítico")
plt.title("Funciones propias - Condiciones Dirichlet")
plt.xlabel("x")
plt.ylabel("X_n(x)")
plt.legend()
plt.grid(True)
plt.show()

# Graficar soluciones numéricas para Neumann
plt.figure(figsize=(12, 6))
for i, sol in enumerate(neumann_solutions):
    plt.plot(x_vals, sol, label=f"n={n_values[i]}, Numérico")
    plt.plot(x_vals, np.cos(n_values[i] * np.pi * x_vals), '--', label=f"n={n_values[i]}, Analítico")
plt.title("Funciones propias - Condiciones Neumann")
plt.xlabel("x")
plt.ylabel("X_n(x)")
plt.legend()
plt.grid(True)
plt.show()
```

```
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_bvp

# Definir el sistema de ecuaciones diferenciales
# y[0] = X, y[1] = X'
def odes(x, y, lambda_n):
    return np.vstack([y[1], -lambda_n * y[0]])

# Condiciones de frontera para Dirichlet
def bc_dirichlet(ya, yb):
    return np.array([ya[0], yb[0]])

# Condiciones de frontera para Neumann
def bc_neumann(ya, yb):
    return np.array([ya[1], yb[1]])

# Valores de x para la solución numérica
x_vals = np.linspace(0, 1, 100)

# Solución numérica para Dirichlet
n_values = [1, 2, 3]
dirichlet_solutions = []
for n in n_values:
    lambda_n = (n * np.pi) ** 2
    y_init = np.zeros((2, x_vals.size))  # Condición inicial aproximada
    sol = solve_bvp(lambda x, y: odes(x, y, lambda_n), bc_dirichlet, x_vals, y_init)
    dirichlet_solutions.append(sol.sol(x_vals)[0])
    print(f"\nSolución para Dirichlet (n={n}):")
    print(sol.sol(x_vals)[0])

# Solución numérica para Neumann
neumann_solutions = []
for n in n_values:
    lambda_n = (n * np.pi) ** 2
    y_init = np.zeros((2, x_vals.size))  # Condición inicial aproximada
    sol = solve_bvp(lambda x, y: odes(x, y, lambda_n), bc_neumann, x_vals, y_init)
    neumann_solutions.append(sol.sol(x_vals)[0])
    print(f"\nSolución para Neumann (n={n}):")
    print(sol.sol(x_vals)[0])

# Graficar soluciones numéricas para Dirichlet
plt.figure(figsize=(12, 6))
for i, sol in enumerate(dirichlet_solutions):
    plt.plot(x_vals, sol, label=f"Numérico (n={n_values[i]})", linewidth=2)
    plt.plot(x_vals, np.sin(n_values[i] * np.pi * x_vals), '--', label=f"Analítico (n={n_values[i]})", linewidth=2)
plt.title("Funciones propias para condiciones de frontera Dirichlet")
plt.xlabel("x")
plt.ylabel("X_n(x)")
plt.legend()
plt.grid(True)
plt.ylim([-1.2, 1.2])
plt.show()

# Graficar soluciones numéricas para Neumann
plt.figure(figsize=(12, 6))
for i, sol in enumerate(neumann_solutions):
    plt.plot(x_vals, sol, label=f"Numérico (n={n_values[i]})", linewidth=2)
    plt.plot(x_vals, np.cos(n_values[i] * np.pi * x_vals), '--', label=f"Analítico (n={n_values[i]})", linewidth=2)
plt.title("Funciones propias para condiciones de frontera Neumann")
plt.xlabel("x")
plt.ylabel("X_n(x)")
plt.legend()
plt.grid(True)
plt.ylim([-1.2, 1.2])
plt.show()
```
