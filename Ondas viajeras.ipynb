```
# CELDA 1: Librerías
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
```

Importamos las librerias

```
# CELDA 2: Parámetros
c = 1      # velocidad de propagación
L = 1      # longitud del dominio
x = np.linspace(0, L, 500)
t_vals = [0, 0.1, 0.2, 0.3, 0.4]
```

Esta celda define las constantes y el dominio sobre el cual se evaluarán las funciones de onda.

```
# CELDA 3: Definición de la función tipo Gauss
def gauss(x, t, direccion='+'):
    if direccion == '+':
        x_prime = x - c * t
    else:
        x_prime = x + c * t
    return np.exp(-((x_prime - 0.5)**2) / 0.01)
```

Esta celda define la forma funcional de la perturbación y cómo esta se propaga

```
# CELDA 4: Graficar ondas viajeras
plt.figure(figsize=(12, 6))

for i, t in enumerate(t_vals):
    plt.subplot(2, 3, i+1)
    f = gauss(x, t, '+')
    g = gauss(x, t, '-')
    u = f + g

    plt.plot(x, f, 'b--', label='Derecha')
    plt.plot(x, g, 'r--', label='Izquierda')
    plt.plot(x, u, 'k', label='Superposición')
    plt.title(f"t = {t:.2f}")
    plt.ylim(0, 2.1)
    plt.grid(True)
    if i == 0:
        plt.legend()

plt.tight_layout()
plt.show()
```

Esta celda es donde se realiza la evaluación numérica y la visualización del principio de superposición.
