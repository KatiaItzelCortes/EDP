### Graficas 3D

```python
from sympy.plotting import plot3d
```

Nos ayuda a graficar en 3 dimenciones

```python
x,y=symbols('x y')
```

Definimos las variables

```python
plot3d(x*y)
```

Crea en grafico(superficie) de la función

```python
plot3d(x*y, (x,-5,5),(y,-5,5))
```

Crea una grafica de la funcion rangos en la variable x y y

```python
plot3d(x*y, -x*y, (x,-5,5),(y,-5,5))
```

Muestra los graficos de dos finciones en un intervalo

```python
plot3d((x**2 + y**2, (x,-5,5),(y,-5,5)),(x*y,(x,-3,3),(y,-3,3)))
```

Crea dos graficos distintos con intervalos diferentes

```python
from sympy.plotting import plot3d_parametric_line
```

Se usa para graficar lineas en 3d

```python
plot3d_parametric_line(cos(s),sin(s),s)
```

Grafica una linea en 3d, cos() da las coordenadas en x, sen() en y y s en z

```python
plot3d_parametric_line((cos(s),sin(s),s,(s,-5,5)),(sin(s),s**2,s,(s,-5,5)))
```

Es el grafico anterior con un intervalo

```python
from sympy.plotting import plot3d_parametric_surface
```

Se usa para graficar superficies en 3d

```python
u, v =symbols('u v')
```

Define dos variables

```python
plot3d_parametric_surface(cos(u+v),sin(u-v),u-v)
```

Grafica una superficie donde cos() da las coordenadas en x, sen() en y y u-v en z

```python
plot3d_parametric_surface(cos(v)*cos(u),cos(v)*sin(u),sin(v))
```

Grafica en 3d y las funciones son las coordenadas en x,y,z

Funciones implícitas

```python
from sympy import plot_implicit, symbols, Eq, And
```

plot_implicit sirve para graficar eciaciones y desigualdades, Eq funcion para crear eciaciones, And función para representar la conjunción logica

```python
x, y = symbols(' x y')
```

Definen las variables

```python
p1 = plot_implicit(Eq(x**2 + y**2, 5))
```

Crea una ecuacion de un circulo y lo grafica

```python
p6 = plot_implicit(y > x**2)
```

Grafica la desigualdad

```python
p6 = plot_implicit(And(y > x, y > -x))
```
